```python
import json
import time
from datetime import datetime, date
import pickle
import os
import random


if __name__ == '__main__':

    log_folder = 'logs'
    try:
        os.mkdir(log_folder)
    except FileExistsError:
        pass

    data_folder = "./data/concept_corpus"
    data_seperate_folder = "data_seperate"

    concept_folder = "concept_seperate"
    try:
        os.mkdir(concept_folder)
    except FileExistsError:
        pass

      # Comment
    concept_list_txt = os.path.join(
        data_folder, 'filtered_keywords_limited_wordlength.txt'
    )

    with open(concept_list_txt, 'r', encoding='utf-8') as f:
        all_concept_lists = [line.strip() for line in f if line.strip()]
      # Comment

    random.seed()
    total_file = 33
    write_file = 0
    cc = 0

    while write_file <= total_file:

        curr_ID = random.randint(0, total_file)
        formatted_ID = '{:02d}'.format(curr_ID)

        data_file = os.path.join(
            data_seperate_folder, f'part_{formatted_ID}.pkl'
        )

        concept_file = os.path.join(
            concept_folder, f'concept_{formatted_ID}.pkl'
        )

        log_file = os.path.join('logs', 'log_' + formatted_ID + '.txt')
        if cc % 10 == 0:
            with open(log_file, 'a', encoding='utf-8') as f:
                f.write(
                    f'formatted_ID: {formatted_ID}; '
                    f'cc: {cc}, write_file num: {write_file}\n'
                )
        cc += 1

        if not os.path.exists(concept_file):

            concepts_for_paper_list = []

            with open(data_file, 'rb') as file:
                paper_info = pickle.load(file)

            concepts_at_least_one = []

            # check all papers
            for id_paper, cur_paper in enumerate(paper_info):

                concepts_for_single_paper = []

                for id_concept, cur_concept in enumerate(all_concept_lists):

                    if cur_concept in cur_paper:
                        concepts_for_single_paper.append(cur_concept)

                concepts_at_least_one.extend(concepts_for_single_paper)

            with open(concept_file, "wb") as output_file:
                pickle.dump(concepts_at_least_one, output_file)
                write_file += 1
```

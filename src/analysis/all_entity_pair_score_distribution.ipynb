# Entity Pair Score Distribution

Data source: `all_predictions_with_entities.csv`

- **Score line chart (descending)**: Trend of entity pair scores from high to low
- **Cumulative score percentage chart**: Cumulative score contribution of top N pairs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set font to avoid Chinese garbled text
plt.rcParams['axes.unicode_minus'] = False

# Read data
df = pd.read_csv('all_predictions_with_entities.csv')
print(f'Total rows: {len(df)}')
print(f'Score range: [{df["score"].min():.6f}, {df["score"].max():.6f}]')
df.head()
```

```python
# Data already sorted by score descending, confirm
print(f'Strictly descending: {df["score"].is_monotonic_decreasing}')

# Add rank column (starting from 1) for position
df = df.reset_index(drop=True)
df['rank'] = range(1, len(df) + 1)
```

```python
# =====================
# Fig 1: Score descending line chart (Quantile view)
# =====================

# Calculate quantile (rank / total)
df['quantile'] = df['rank'] / len(df)

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(df['quantile'], df['score'], linewidth=0.6, color='#2196F3', alpha=0.8)
ax.fill_between(df['quantile'], df['score'], alpha=0.1, color='#2196F3')

# Mark key statistics
mean_score = df['score'].mean()
median_score = df['score'].median()
ax.axhline(y=mean_score, color='#FF5722', linestyle='--', linewidth=1.2, label=f'Mean = {mean_score:.6f}')
ax.axhline(y=median_score, color='#4CAF50', linestyle=':', linewidth=1.2, label=f'Median = {median_score:.6f}')

ax.set_xlabel('Quantile', fontsize=12)
ax.set_ylabel('Score', fontsize=12)

ax.legend(fontsize=10)
ax.set_xlim(0, 1)
ax.set_ylim(0.35, 0.55)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f'Score stats: max={df["score"].max():.6f}, min={df["score"].min():.6f}')
print(f'Mean={mean_score:.6f}, Median={median_score:.6f}, Std={df["score"].std():.6f}')
```
